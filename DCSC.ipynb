```python
import pandas as pd

df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding='latin-1')
```

```python
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
df.info()
```

```python
# change the column name form Upper with spaces to lower case with underscore (snake_casing)

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns={'type' : 'payment_type' , 
                        'shipping_date_(dateorders)' : 'shipping_date' , 
                        'days_for_shipping_(real)' : 'actual_shipping_days', 
                        'days_for_shipment_(scheduled)' : 'scheduled_shipping_days' , 
                        'order_date_(dateorders)' : 'order_date'})
```

```python
df.head(5)
```

```python
df.columns
```

```python
#Checking empty column
df.isnull().sum()
```

```python
#Droping columns not usable for analysis and dashboard
df = df.drop(columns=['order_zipcode' , 'product_description' , 'customer_email' , 'customer_password' , 'customer_zipcode']) 
```

```python
df.isnull().sum()
```

```python
#checking number of duplicate rows

num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows found: {num_duplicates}")
```

```python
#Standardising date columns in date format
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
df['shipping_date'] = pd.to_datetime(df['shipping_date'], format='mixed')
```

```python
# Validating actual shipping days column 
df['calculated_duration'] = (df['shipping_date'] - df['order_date']).dt.days
```

```python
mask_inconsistent = df['calculated_duration'] != df['actual_shipping_days']
bad_data = df[mask_inconsistent]
```

```python
print(f"Rows with shipping discrepancies: {len(bad_data)}")
print(bad_data[['order_date', 'shipping_date', 'calculated_duration', 'actual_shipping_days']].head())
```

```python
# New column for delay in shipment
df['delay_variance'] = df['actual_shipping_days'] - df['scheduled_shipping_days']
print(df[['scheduled_shipping_days', 'actual_shipping_days', 'delay_variance']].head())
```

```python
#Standardize categorical text
cols_to_fix = ['order_city', 'order_country', 'order_region', 'order_state' ,'market']
for col in cols_to_fix:
    df[col] = df[col].str.lower().str.strip().str.title()
```

```python
from sqlalchemy import create_engine

#Connecting to PostgreSQL
username = "postgres"      # default user
password = "admin" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "DCSC"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#Loading DataFrame into PostgreSQL
table_name = "supply_chain"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
